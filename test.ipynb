```python
# Ввод данных
N, M = map(int, input().split())
weight = [list(map(int, input().split())) for _ in range(N)]

# Инициализация dp массива
dp = [[float('inf')] * M for _ in range(N)]
dp[0][0] = weight[0][0]  # Начальная точка

# Заполнение dp массива
for i in range(N):
    for j in range(M):
        if i > 0:
            dp[i][j] = min(dp[i][j], dp[i-1][j] + weight[i][j])
        if j > 0:
            dp[i][j] = min(dp[i][j], dp[i][j-1] + weight[i][j])

# Ответ - минимальное количество еды для достижения правого нижнего угла
print(dp[N-1][M-1])
```

```python
import random
from faker import Faker

# Инициализация Faker
fake = Faker()

# Генерация данных для Authors
def generate_authors(num_authors):
    authors = []
    for _ in range(num_authors):
        author = (
            _ + 1,  # author_id
            fake.name(),  # name
            random.randint(1800, 2000),  # birth_year
            fake.country()  # nationality
        )
        authors.append(author)
    return authors

# Генерация данных для Books
def generate_books(num_books, num_authors):
    books = []
    for _ in range(num_books):
        book = (
            _ + 1,  # book_id
            fake.catch_phrase(),  # title
            random.randint(1, num_authors),  # author_id
            random.randint(1800, 2024),  # published_year
            fake.word(),  # genre
            random.randint(1, 10)  # available_copies
        )
        books.append(book)
    return books

# Генерация данных для Users
def generate_users(num_users):
    users = []
    for _ in range(num_users):
        user = (
            _ + 1,  # user_id
            fake.name(),  # name
            fake.email(),  # email
            fake.date_between(start_date='-2y', end_date='today')  # registration_date
        )
        users.append(user)
    return users

# Генерация данных для Loans
def generate_loans(num_loans, num_books, num_users):
    loans = []
    for _ in range(num_loans):
        loan = (
            _ + 1,  # loan_id
            random.randint(1, num_books),  # book_id
            random.randint(1, num_users),  # user_id
            fake.date_between(start_date='-1y', end_date='today'),  # loan_date
            fake.date_between(start_date='today', end_date='+30d')  # return_date
        )
        loans.append(loan)
    return loans

# Настройки генерации данных
num_authors = 10
num_books = 50
num_users = 20
num_loans = 30

# Генерация данных
authors = generate_authors(num_authors)
books = generate_books(num_books, num_authors)
users = generate_users(num_users)
loans = generate_loans(num_loans, num_books, num_users)

# Вывод сгенерированных данных в нужном формате
print("Authors:")
for author in authors:
    print(author)

print("\nBooks:")
for book in books:
    print(book)

print("\nUsers:")
for user in users:
    print(user)

print("\nLoans:")
for loan in loans:
    print(loan)
```
